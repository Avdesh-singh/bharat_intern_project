```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer as tf
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
dir(tf)
```

```python

df = pd.read_csv('spam.csv')
df
```

```python
df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace = True)
df
```

```python
df.rename(columns={'v1':'label', 'v2':'message'}, inplace = True)
df
```

```python
# Encode the labels
encoder = LabelEncoder()
df['label'] = encoder.fit_transform(df['label'])
df['label']
```

```python

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42)
```

```python
# Transform text data to feature vectors
vectorizer = tf( max_features=3000)
X_train_vec= vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)



```

```python

# Build the model
model = Sequential()
model.add(Dense(512, input_dim=X_train_vec.shape[1], activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python

# Train the model
history = model.fit(X_train_vec.toarray(), y_train, epochs=10, batch_size=64, validation_data=(X_test_vec.toarray(), y_test))
```

```python

# Save the model
model.save('sms_spam_model.h5')
```
